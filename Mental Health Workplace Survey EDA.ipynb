## Visualizations - Univariate/Bivariate/Multivariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: BurnoutLevel by Age Group and Gender
df['AgeGroup'] = pd.cut(df['Age'], bins=[20, 30, 40, 50, 60], labels=['21â€“30', '31â€“40', '41â€“50', '51â€“60'])
plot1 = sns.barplot(ax=axes[0], x='AgeGroup', y='BurnoutLevel', hue='Gender', data=df, ci=None)
axes[0].set_title('Burnout Level by Age Group and Gender')
axes[0].set_xlabel('Age Group')
axes[0].set_ylabel('Average Burnout Level')
plot1.legend(title='Gender')
for container in plot1.containers:
    plot1.bar_label(container, fmt='%.2f', padding=3)

# Plot 2: Average BurnoutLevel by Country (Top 10)
top_countries = df['Country'].value_counts().nlargest(10).index
country_subset = df[df['Country'].isin(top_countries)]
plot2 = sns.barplot(ax=axes[1], x='Country', y='BurnoutLevel', data=country_subset, estimator=np.mean, ci=None)
axes[1].set_title('Burnout Level by Country (Top 10)')
axes[1].set_xlabel('Country')
axes[1].set_ylabel('Average Burnout Level')
axes[1].tick_params(axis='x', rotation=45)
for container in plot2.containers:
    plot2.bar_label(container, fmt='%.2f', padding=3)
    
# Final layout adjustment
plt.tight_layout()
plt.show()
```

```python
# Create 1x3 layout
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot 1: Distribution of Burnout Level (histogram with labels)
histplot = sns.histplot(ax=axes[0], data=df, x='BurnoutLevel', bins=20, kde=True)
axes[0].set_title('Distribution of Burnout Level')
axes[0].set_xlabel('Burnout Level')
axes[0].set_ylabel('Frequency')
# Add data labels to histogram bars
for container in histplot.containers:
    histplot.bar_label(container, fmt='%d', padding=2)

# Plot 2: Work Hours vs Burnout Risk (boxplot)
sns.boxplot(ax=axes[1], x='BurnoutRisk', y='WorkHoursPerWeek', data=df)
axes[1].set_title('Work Hours vs Burnout Risk')
axes[1].set_xlabel('Burnout Risk')
axes[1].set_ylabel('Work Hours Per Week')

# Plot 3: Sleep Hours vs Burnout Risk (boxplot)
sns.boxplot(ax=axes[2], x='BurnoutRisk', y='SleepHours', data=df)
axes[2].set_title('Sleep Hours vs Burnout Risk')
axes[2].set_xlabel('Burnout Risk')
axes[2].set_ylabel('Sleep Hours')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: HasTherapyAccess 
barplot = sns.barplot(ax=axes[0, 0], x='HasTherapyAccess', y='BurnoutRisk', data=df, estimator=np.mean, ci=None)
axes[0, 0].set_title('Burnout Risk by Therapy Access')
axes[0, 0].set_xlabel('Has Therapy Access')
axes[0, 0].set_ylabel('Average Burnout Risk')
barplot.bar_label(barplot.containers[0], fmt='%.2f')

# Plot 2: ManagerSupportScore 
sns.boxplot(ax=axes[0, 1], x='BurnoutRisk', y='ManagerSupportScore', data=df)
axes[0, 1].set_title('Manager Support Score by Burnout Risk')
axes[0, 1].set_xlabel('Burnout Risk')
axes[0, 1].set_ylabel('Support Score')
means = df.groupby('BurnoutRisk')['ManagerSupportScore'].mean()
for i, val in enumerate(means):
    axes[0, 1].text(i, val + 0.3, f'{val:.2f}', ha='center')

# Plot 3: MentalHealthDaysOff 
sns.boxplot(ax=axes[1, 0], x='BurnoutRisk', y='MentalHealthDaysOff', data=df)
axes[1, 0].set_title('Mental Health Days Off by Burnout Risk')
axes[1, 0].set_xlabel('Burnout Risk')
axes[1, 0].set_ylabel('Days Off')
means = df.groupby('BurnoutRisk')['MentalHealthDaysOff'].mean()
for i, val in enumerate(means):
    axes[1, 0].text(i, val + 0.3, f'{val:.2f}', ha='center')

# Plot 4: WorkLifeBalanceScore 
sns.boxplot(ax=axes[1, 1], x='BurnoutRisk', y='WorkLifeBalanceScore', data=df)
axes[1, 1].set_title('Work-Life Balance Score by Burnout Risk')
axes[1, 1].set_xlabel('Burnout Risk')
axes[1, 1].set_ylabel('Balance Score')
means = df.groupby('BurnoutRisk')['WorkLifeBalanceScore'].mean()
for i, val in enumerate(means):
    axes[1, 1].text(i, val + 0.3, f'{val:.2f}', ha='center')

plt.tight_layout()
plt.show()
```

## Subgroup Burnout Trends

## Outlier Detection using Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_columns = ['WorkHoursPerWeek', 'BurnoutLevel', 'StressLevel', 'SleepHours', 'ProductivityScore']

# Plot settings
n_cols = 2
n_rows = (len(numeric_columns) + 1) // 2
fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 12))  # Increased size

# Flatten axes for easy indexing
axes = axes.flatten()

# Generate vertical boxplots with mean labels
for i, col in enumerate(numeric_columns):
    sns.boxplot(ax=axes[i], y=df[col])
    axes[i].set_title(f'Boxplot of {col}')
    axes[i].set_ylabel(col)
    
    # Add mean annotation
    mean_val = df[col].mean()
    axes[i].text(0, mean_val + 0.1, f'Mean: {mean_val:.2f}', 
                 ha='center', va='bottom', color='black', fontsize=9)

# Hide unused subplot if the number of plots is odd
if len(numeric_columns) % 2 != 0:
    axes[-1].axis('off')

plt.tight_layout()
plt.show()
```

# Summary 

This exploratory data analysis was conducted to identify how workplace and lifestyle conditions influence burnout levels and mental health risk across individuals. Below is a structured summary of the analysis:

## 1. Data Overview
- Dataset contains 3,000 records and 25 columns with no missing values.
- Includes demographics, work conditions, support systems, and mental health indicators.

## 2. Descriptive Statistics
- Key numeric variables like Work Hours, Stress Level, Sleep Hours, and Burnout Level were summarized.
- Categorical features such as Gender, Country, and RemoteWork were explored through counts.

## 3. Visualizations and Relationships
- Burnout Level distributions were explored across age, gender, and country.
- Bivariate analysis showed that higher burnout is associated with:
  - Fewer sleep hours
  - Higher work hours
  - Poorer work-life balance
  - Lack of access to therapy or low manager support

## 4. Subgroup Trends
- Department, Salary Range, and Remote Work mode showed distinct burnout patterns.
- Clear class imbalance in BurnoutRisk was noted.

## 5. Correlation Insights
- BurnoutLevel is strongly correlated with StressLevel and WorkHours.
- SleepHours and WorkLifeBalance show negative correlations with BurnoutRisk.

## 6. Outlier Detection
- Outliers were detected in StressLevel, WorkHours, and ProductivityScore.
